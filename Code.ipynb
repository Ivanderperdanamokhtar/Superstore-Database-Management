```python
#Menampilkan konsumen yang melakukan pembelian dengan kuantitas tertinggi:
```

```python
%%sql
SELECT Konsumen.Nama_Konsumen, MAX(Produk.Kuantitas) AS Kuantitas_Tertinggi
FROM Konsumen
JOIN Superstore ON Konsumen.ID_Konsumen = Superstore.ID_Konsumen
JOIN Produk ON Superstore.ID_Produk = Produk.ID_Produk
GROUP BY Konsumen.Nama_Konsumen;
```

```python
#Menampilkan konsumen yang melakukan pembelian dengan kuantitas terendah:
```

```python
%%sql
SELECT Konsumen.Nama_Konsumen, MIN(Produk.Kuantitas) AS Kuantitas_Terendah
FROM Konsumen
JOIN Superstore ON Konsumen.ID_Konsumen = Superstore.ID_Konsumen
JOIN Produk ON Superstore.ID_Produk = Produk.ID_Produk
GROUP BY Konsumen.Nama_Konsumen;
```

```python
#mencari konsumen yang melakukan pembelian dengan profit lebih dari 100:
```

```python
%%sql
SELECT Konsumen.Nama_Konsumen, SUM(Produk.Profit) AS Total_Profit
FROM Konsumen
JOIN Superstore ON Konsumen.ID_Konsumen = Superstore.ID_Konsumen
JOIN Produk ON Superstore.ID_Produk = Produk.ID_Produk
GROUP BY Konsumen.Nama_Konsumen
HAVING Total_Profit > 100;
```
